```python
np.var(nums)
```

```python
np.mean(nums)
```

```python
def var(nums, pop=True):
    mean = sum(nums)/len(nums)
    if pop:
        return sum((nums - mean)**2) / (len(nums) - 1)
    return sum((nums - mean)**2) / len(nums)
```

```python
var(nums, pop=False)
```

```python
def std(var):
    return var**.5
```

```python
std(var(nums, pop=True))
```

```python
x_axis = np.arange(0, 250, 1)

plt.plot(x_axis, norm.pdf(x_axis, np.mean(x_axis), np.std(x_axis)))
plt.plot(x_axis, cauchy.pdf(x_axis, np.mean(x_axis), np.std(x_axis)))

plt.show();
```

```python
a = 1

gamma_vals = np.linspace(gamma.ppf(0.01, a),
                         gamma.ppf(0.99, a), 100)

plt.plot(gamma_vals, gamma.pdf(gamma_vals, a))

plt.show();
```

```python
m, b = np.polyfit(data['Height'], data['Weight'], 1)

fig, axes = plt.subplots(2,1, tight_layout=True, figsize=(10,6))

axes[0].scatter(data['Height'], data['Weight'])
axes[0].plot(data['Height'], m*data['Height'] + b, color='g')

axes[1].scatter(data.loc[(data['Gender']=='Male'), 'Height'], data.loc[(data['Gender']=='Male'), 'Weight'], alpha=0.5)

axes[1].scatter(data.loc[(data['Gender']=='Female'), 'Height'], data.loc[(data['Gender']=='Female'), 'Weight'], alpha=0.5)

plt.show();
```
